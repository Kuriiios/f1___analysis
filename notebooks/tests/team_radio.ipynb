```python
import requests
from bs4 import BeautifulSoup
import re

def scrape_team_radio(url):
    res = requests.get(url)
    soup = BeautifulSoup(res.content, "html.parser")

    # Extract main article content
    article = soup.find("div", class_="article-content")
    if not article:
        print("No article content found.")
        return []

    raw_lines = article.get_text(separator="\n").split("\n")
    transcript = []

    # Extract radio lines
    for line in raw_lines:
        line = line.strip()
        if not line or ":" not in line:
            continue  # Skip empty or non-dialogue lines

        match = re.match(r"^(.*?):\s*(.*)$", line)
        if match:
            speaker = match.group(1).strip()
            message = match.group(2).strip()
            transcript.append({
                "Speaker": speaker,
                "Message": message
            })

    return transcript

# Example use:
url = "https://www.racefans.net/2025/07/01/i-need-some-pace-please-help-full-transcript-of-norris-and-piastris-austrian-gp-battle/"
radio_data = scrape_team_radio(url)

for line in radio_data:
    print(f"{line['Speaker']}: {line['Message']}")
```
